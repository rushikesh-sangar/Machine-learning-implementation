# Lasso & Ridge regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.metrics import r2_score
```

```python
data = pd.read_csv(r'D:\Full Stack Data Science With Gen AI & Agentic AI Notes\Notes\May\12 May\Lasso, ridge & elastic net\Task - Lasso & ridge\Car mpg.csv')
```

```python
data.info()
```

```python
data.head()
```

```python
data = data.drop(['car_name'], axis = 1)
data['origin'] = data['origin'].replace({1:'america', 2:'europe', 3:'asia'})
data = pd.get_dummies(data, columns = ['origin'], dtype = int)
```

```python
data.head()
```

```python
data = data.replace('?', np.nan)
data = data.apply(pd.to_numeric)
data = data.apply(lambda x : x.fillna(x.median()) if x.dtype != 'object' else x, axis = 0)
```

```python
data.head()
```

```python
x = data.drop(['mpg'], axis = 1)
y = data[['mpg']]
```

```python
x.head()
```

```python
 y.head()
```

```python
x_s = preprocessing.scale(x)
x_s = pd.DataFrame(x_s, columns = x.columns)
```

```python
x_s
```

```python
y_s = preprocessing.scale(y)
y_s = pd.DataFrame(y_s, columns = y.columns)
```

```python
y_s
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_s, y_s, test_size = 0.3, random_state = 0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
len(data)
```

```python
data.columns
```

```python
regression_model = LinearRegression()
regression_model.fit(x_train, y_train)

for idx, col_name in enumerate(x_train.columns):
    print('The coef for {} is {}'.format(col_name, regression_model.coef_[0][idx]))

intercept = regression_model.intercept_[0]
print('The intercept is {}'.format(intercept))
```

```python
# Ridge model

ridge_model = Ridge(alpha = 0.4)
ridge_model.fit(x_train, y_train)
print('Ridge model coef: {}'.format(ridge_model.coef_))
```

```python
# Lasso model

lasso_model = Lasso(alpha = 0.1)
lasso_model.fit(x_train, y_train)
print('Lasso model coef:{}'.format(lasso_model.coef_))
```

```python
# Model score - R-squared

# Simple linear model
print(regression_model.score(x_train, y_train))
print(regression_model.score(x_test, y_test))
print('******************')

# Ridge
print(ridge_model.score(x_train, y_train))
print(ridge_model.score(x_test, y_test))
print('******************')

# Lasso
print(lasso_model.score(x_train, y_train))
print(lasso_model.score(x_test, y_test))
```
